```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv('RAW_recipes.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
df = df[['id', 'name', 'ingredients', 'tags', 'steps', 'minutes' ]]
```

```python
df.head()
```

```python
df['ingredients'] = df['ingredients'].apply(ast.literal_eval)
```

```python
df['ingredients'].iloc[0]
```

```python
df['tags'] = df['tags'].apply(ast.literal_eval)
```

```python
df['tags'].iloc[0]
```

```python
def clean_tokens(token_list):
    return [i.replace(" ", "").lower() for i in token_list]
```

```python
df['ingredients_clean'] = df['ingredients'].apply(clean_tokens)

df['tags_clean'] = df['tags'].apply(clean_tokens)
```

```python
df[['ingredients_clean', 'tags_clean' ]].head()
```

```python
df['features'] = df['ingredients_clean'] + df['tags_clean']
```

```python
df['features'] = df['features'].apply(lambda x: " ".join(x))
```

```python
df[['name', 'features']].head()
```

```python
df['name'] = df['name'].fillna('')
```

```python
df[df['name'].str.contains("pizza", case = False)].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
```

```python
tfidf = TfidfVectorizer(stop_words = 'english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['features'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
indices = pd.Series(df.index, index = df['name']).drop_duplicates()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend_recipes(recipe_name, top_n=5):

    recipe_name = recipe_name.lower()

    matches = df[df['name'].str.lower().str.contains(recipe_name, na=False)]

    if matches.empty:
        return "Recipe not found"

    idx = matches.index[0]

    recipe_vector = tfidf_matrix[idx]

    similarity_scores = cosine_similarity(recipe_vector, tfidf_matrix).flatten()

    similar_indices = similarity_scores.argsort()[-top_n-1:-1][::-1]

    return df['name'].iloc[similar_indices].reset_index(drop=True)
```

```python
def recommend_by_ingredients(ingredients, top_n=5):

    ingredients = ingredients.lower()

    query_vector = tfidf.transform([ingredients])

    similarity_scores = cosine_similarity(query_vector, tfidf_matrix).flatten()

    similar_indices = similarity_scores.argsort()[-top_n:][::-1]

    return df['name'].iloc[similar_indices].reset_index(drop=True)
```

```python
recommend_recipes("pizza")
```

```python
recommend_recipes("chicken")
```

```python
recommend_recipes("chili")
```

```python
recommend_by_ingredients("chicken tomato garlic")
```

```python
recommend_by_ingredients("egg cheese milk")
```
